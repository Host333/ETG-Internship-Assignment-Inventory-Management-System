# Purchase

```python
import json
```

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

Inventory = json.loads(r)
```

```python
Inventory
```

```python
import time
```

```python
t = time.ctime()
ui_prod  = input("Enter the Product_id: ")
ui_quant = int(input("Enter the quantity you want to purchase: "))

if ui_quant <= Inventory[ui_prod]['Quantity']:
    print('----------------------------------------------------')
    print("You ordered: ", Inventory[ui_prod]['Name'])
    print("Price/unit: ", Inventory[ui_prod]['Price'])
    print("Quantity ordered: ", ui_quant)
    print('====================================================')

    print("Your Billing Amount is: ", Inventory[ui_prod]['Price'] * ui_quant)
    print('====================================================')

    print("Thankyou for shopping with us. Visit againðŸ˜€\n\n")
    print("Transaction Date & Time: ", t)
    print('----------------------------------------------------')
    Inventory[ui_prod]['Quantity'] = Inventory[ui_prod]['Quantity'] - ui_quant
else:
    print('\n----------------------------------------------------')
    print("Sorry! we are only left with", Inventory[ui_prod]['Quantity'], "products.")

    print("Do you wish to buy the quantity available?")

    ui_ans = input(" Type Y or N: ")

    if ui_ans == 'Y':
        print('\n----------------------------------------------------')
        print("You ordered: ", Inventory[ui_prod]['Name'])
        print("Price/unit: ", Inventory[ui_prod]['Price'])
        print("Quantity ordered: ", Inventory[ui_prod]['Quantity'])
        print('====================================================')

        print("Your Billing Amount is: ", Inventory[ui_prod]['Price'] * Inventory[ui_prod]['Quantity'])
        print('====================================================')

        print("Thankyou for shopping with us. Visit againðŸ˜€\n\n")
        print("Transaction Date & Time: ", t)
        print('----------------------------------------------------')

        Inventory[ui_prod]['Quantity'] = Inventory[ui_prod]['Quantity'] - Inventory[ui_prod]['Quantity']
    
    else:
        print('----------------------------------------------------')
        print("\nThank-you! Please visit againðŸ˜€")


```

# Inventory Update 

```python
js = json.dumps(Inventory)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

# Sales JSON file

```python
sale = {}
```

```python
fds = open('sales.json', 'w')
sale[len(sale) + 1] = {'Product ID': ui_prod, 'Product': Inventory[ui_prod]['Name'], 'Quantity': ui_quant, 'Amount': ui_quant * Inventory[ui_prod]['Price'], 'Transaction Date & Time': t}

sales = json.dumps(sale)
fds.write(sales)

fds.close()
```
